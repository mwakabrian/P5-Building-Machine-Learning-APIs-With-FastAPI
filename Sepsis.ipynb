*These values help us understand the relative importance of each principal component in capturing the variability of the original data.*

**Cumulative Explained Variance Ratio:**

*Each value in the cumulative explained variance ratio represents the cumulative proportion of variance explained by including all principal components up to that point.*
*For instance, the cumulative explained variance after considering the first principal component is approximately 24.67%, after including the second principal component, it becomes approximately 42.33%, and so forth.*
*These values indicate how much of the total variability in the data is captured when considering a certain number of principal components.*

*The final value in the cumulative explained variance ratio is always 1, indicating that all variance in the data is accounted for when considering all principal components*

*The explained variance ratio helps us understand the contribution of each principal component towards capturing the variability in the data. Components with higher ratios are more influential in explaining the data's structure.*

*The cumulative explained variance ratio helps us decide how many principal components to retain for dimensionality reduction or data compression purposes. We typically aim to retain a sufficient number of components that capture a high proportion of the total variance, while discarding components with low explanatory power.*

*In this specific example, we can see that the first few principal components explain a significant portion of the variance in the data. By including the first five or six components, we capture approximately 76-84% of the total variance, which might be sufficient for many practical purposes.*

### **HYPOTHESIS TESTING** ###

**Null Hypothesis (H0):**
*There is no significant difference in Blood_Pressure and BMI between patients with and without sepsis.*

**Alternative Hypothesis (H1):**
*There is a significant difference in Blood_Pressure and BMI between patients with and without sepsis.*

*Split the data into sepsis and non-sepsis groups:*

```python
# Split the data into sepsis and non-sepsis groups
sepsis_group = train_data[train_data['Target_Sepsis'] == True]
non_sepsis_group = train_data[train_data['Target_Sepsis'] == False]
```

**Perform statistical tests:**

*We use independent two-sample t-tests to compare the means of Blood_Pressure and BMI between the sepsis and non-sepsis groups. However, before applying the t-test, we need to check if the data meets the assumptions of normality and homogeneity of variances.*

**Performing a test for normality assumption using the Shapiro-Wilk test for Blood Pressure and BMI**

```python
import scipy.stats as stats

# Check normality assumption
_, sepsis_pressure_norm = stats.shapiro(sepsis_group['Blood_Pressure'])
_, non_sepsis_pressure_norm = stats.shapiro(non_sepsis_group['Blood_Pressure'])
_, sepsis_bmi_norm = stats.shapiro(sepsis_group['BMI'])
_, non_sepsis_bmi_norm = stats.shapiro(non_sepsis_group['BMI'])
```

**Checking the homogeneity of variances between two groups for Blood Pressure and BMI.**

```python
# Check homogeneity of variances
_, pressure_var_equal = stats.levene(sepsis_group['Blood_Pressure'], non_sepsis_group['Blood_Pressure'])
_, bmi_var_equal = stats.levene(sepsis_group['BMI'], non_sepsis_group['BMI'])
```

**Performing independent samples t-tests for Blood Pressure and BMI, between the sepsis group and the non-sepsis group**

```python
# Perform t-tests
if sepsis_pressure_norm > 0.05 and non_sepsis_pressure_norm > 0.05 and pressure_var_equal > 0.05:
    t_stat_pressure, p_val_pressure = stats.ttest_ind(sepsis_group['Blood_Pressure'], non_sepsis_group['Blood_Pressure'])
else:
    t_stat_pressure, p_val_pressure = stats.mannwhitneyu(sepsis_group['Blood_Pressure'], non_sepsis_group['Blood_Pressure'])

if sepsis_bmi_norm > 0.05 and non_sepsis_bmi_norm > 0.05 and bmi_var_equal > 0.05:
    t_stat_bmi, p_val_bmi = stats.ttest_ind(sepsis_group['BMI'], non_sepsis_group['BMI'])
else:
    t_stat_bmi, p_val_bmi = stats.mannwhitneyu(sepsis_group['BMI'], non_sepsis_group['BMI'])
```

```python
# Print the results
print(f"Blood Pressure Test Results:")
print(f"t-statistic: {t_stat_pressure:.4f}, p-value: {p_val_pressure:.4f}")
print(f"\nBMI Test Results:")
print(f"t-statistic: {t_stat_bmi:.4f}, p-value: {p_val_bmi:.4f}")
```

**Based on the test results:**

1. **For Blood Pressure:**
   - *The t-statistic is 49100.5000.*
   - *The p-value is 0.0000, indicating that it is extremely unlikely to observe such extreme results under the null hypothesis.*
   - *With such a low p-value, we reject the null hypothesis and conclude that there is a significant difference in Blood Pressure between patients with and without sepsis.*
